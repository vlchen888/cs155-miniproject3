```python
#nltk.download('cmudict')
from nltk.corpus import cmudict
d = cmudict.dict()
def nsyl(word):
  return [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]] 


shake_words = [] 
with open('data/Syllable_dictionary.txt') as f:
    for line in f:
        arr = line.split()
        word = arr[0]
        shake_words.append(word)

extra_words = []
import re

for s in list(sonnets.keys()):
    for line in sonnets[s]:
        for word in line:
            clean_word = re.sub(r'[^\w\'\-]', '', word).lower()
            if (clean_word not in extra_words) and (clean_word not in shake_words):
                extra_words.append(clean_word)

extra_words_int = []                
# Get interior syllable counts for each extra word                
extra_words.sort()
manual_words = []
for j in range(len(extra_words)):
    try:
        #print(extra_words[j]+' %i'% nsyl(extra_words[j])[0])
        nsyl_test = nsyl(extra_words[j])[0]
        extra_words_int.append(extra_words[j])
        extra_words_int.append(nsyl_test)
    except KeyError:
        manual_words.append(extra_words[j])
        pass

shake_words_int = []
with open('data/Syllable_dictionary.txt') as f:
    for line in f:
        arr = line.split()
        word = arr[0]
        shake_words_int.append(word)
        for i in range(1, len(arr)):
            token = arr[i]
            if token[0] != 'E':
                if len(shake_words_int)%2 == 1:
                    shake_words_int.append(token)
                else:
                    shake_words_int[-1] = token

extra_words_int = extra_words_int + manual_syll
all_words_int = extra_words_int + shake_words_int

d_int = {}
for s in range(int(len(extra_words_int)/2)):
    d_int[extra_words_int[2*s]] = int(extra_words_int[2*s+1])
    
d_all = {}
for s in range(int(len(all_words_int)/2)):
    d_all[all_words_int[2*s]] = int(all_words_int[2*s+1])
    
d_ext = dict(d_int)
# Get end word syllable counts
for s in list(sonnets.keys()):
#for s in range(1,2):
    for line in sonnets[s]:
        lsum = 0
        for j in range(0,len(line)-1):
            clean_word = re.sub(r'[^\w\'\-]', '', line[j]).lower()
            lsum += d_all[clean_word]
                
        clean_word = re.sub(r'[^\w\'\-]', '', line[-1]).lower()
        try:
            if d_int[clean_word] < 10-lsum:            
                d_ext[clean_word] = '%i E%i'%(d_int[clean_word],10-lsum)
            elif (d_int[clean_word] > 10-lsum) and (lsum < 10):
                d_ext[clean_word] = 'E%i %i'%(10-lsum,d_int[clean_word])
        except KeyError:
            pass
        except TypeError:
            print(d_int[clean_word])
 
# Construct the full dictionary
s_dict = {}
with open('data/Syllable_dictionary.txt') as f:
    for line in f:
        arr = line.split()
        syllables = ''
        for i in range(1, len(arr)):
            syllables = syllables + ' ' + arr[i]
        s_dict[arr[0]] = syllables

for s in list(d_ext.keys()):
    s_dict[s] = ' ' + str(d_ext[s])
     
    
# Output to file
open('syllable_dict_expanded.txt', 'w').close()
f = open('syllable_dict_expanded.txt', "a")
for key in sorted(s_dict):
    f.write('%s%s\n' % (key, s_dict[key]))
f.close()
                
```
